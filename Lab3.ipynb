```python
data_set = pd.read_csv('dataset_lab3.csv', na_values='?')
data_set.columns = feature_names
data_set['normalized-losses'] = pd.to_numeric(data_set['normalized-losses'], errors='coerce')
data_set.head()

```

```python
# print toal number of values  min , max and standard deviation for normalized-losses
print(data_set['normalized-losses'].count())
print(data_set['normalized-losses'].min())
print(data_set['normalized-losses'].max())
print(data_set['normalized-losses'].std())




```

```python
# Find the 25% and 75% percentile – you can use the function .quantile()
print(data_set['normalized-losses'].quantile(0.25))
print(data_set['normalized-losses'].quantile(0.75))
data_set['normalized-losses'].describe()

```

```python
# For the feature make print the total number of values and the list of unique values.You can use .unique() function.
print(data_set['make'].count())
print(data_set['make'].unique())
```

```python
# explore the mode() function to get the name and value of the mode
print(data_set['make'].mode())
```

```python
# Explore the value_counts() function to get the mode and second mode
print(data_set['make'].value_counts().head(2))
```

```python

# Create the report row for normalized-losses as shown above. The values should be calculated using the functions we looked at today, not hard-coded.

featrue = 'normalized-losses'

report_row = {
    'feature': featrue,
    'count': data_set[featrue].count(),
    'min': data_set[featrue].min(),
    'max': data_set[featrue].max(),
    'std': data_set[featrue].std(),
    '25%': data_set[featrue].quantile(0.25),
    '75%': data_set[featrue].quantile(0.75),
    'mode': data_set[featrue].mode().values[0],
    'mode_count': data_set[featrue].value_counts().max()
}


report = pd.DataFrame([report_row])

report.reset_index(drop=True, inplace=True)

report.columns

report

```

```python

all_rows = []


report = pd.DataFrame(columns=['feature', 'count', 'min', 'max', 'std', '25%', '75%'])

for feature in data_set.columns:
    if pd.api.types.is_numeric_dtype(data_set[feature]):

        stats = {
            'Feature': feature,
            'Count': data_set[feature].count(),
            'missing_values': data_set[feature].isnull().sum(),
            'Min': data_set[feature].min(),
            'Max': data_set[feature].max(),
            'Std': data_set[feature].std(),
            '25%': data_set[feature].quantile(0.25),
            '75%': data_set[feature].quantile(0.75)
        }
        all_rows.append(stats)

report = pd.DataFrame(all_rows)


style_report = report.style.hide(axis="index")


html = style_report.to_html()

html

from IPython.core.display import HTML
display(HTML(html))




```
